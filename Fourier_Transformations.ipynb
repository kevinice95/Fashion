```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
percent = pd.read_csv('fictionfabrics%.csv')
```

```python
# remove columns with only 0s and remove first. get years vector
years = percent.iloc[:,0]
percent = percent.iloc[:,2:]
percent = percent.loc[:, (percent != 0).any(axis=0)]
```

```python
for i in range(0,len(percent.columns)-1):
    a = np.array(percent.iloc[:,i])
    a = (a - np.mean(a))/np.std(a)
    #plt.plot(range(0,209),a)
    a = np.fft.rfft(a)
    plt.plot(range(0,105),a)
```

```python
for i in range(0,len(percent.columns)-1):
    a = np.array(percent.iloc[:,i])
    a = (a - np.mean(a))/np.std(a)
    #plt.plot(range(0,209),a)
    a = np.correlate(a,a,mode='full')[len(a)-1:]
    plt.plot(range(0,209),a)
```
